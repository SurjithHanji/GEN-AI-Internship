```python
# Q19 - Text Analysis Functions

# count words
def count_words(t):
    words = t.split()
    return len(words)

# count vowels
def count_vowels(t):
    count = 0
    for ch in t.lower():
        if ch in "aeiou":
            count = count + 1
    return count

# count consonants
def count_consonants(t):
    count = 0
    for ch in t.lower():
        if ch.isalpha() and ch not in "aeiou":
            count = count + 1
    return count

# reverse text
def reverse_text(t):
    return t[::-1]

# check palindrome
def check_palindrome(t):
    text = t.lower().replace(" ", "")
    if text == text[::-1]:
        return True
    else:
        return False

# remove vowels
def remove_vowels(t):
    new_text = ""
    for ch in t:
        if ch.lower() not in "aeiou":
            new_text = new_text + ch
    return new_text

# find longest word
def longest_word(t):
    words = t.split()
    long_word = ""
    for w in words:
        if len(w) > len(long_word):
            long_word = w
    return long_word

# word frequency
def word_frequency(t):
    words = t.lower().split()
    freq = {}

    for w in words:
        if w in freq:
            freq[w] = freq[w] + 1
        else:
            freq[w] = 1

    return freq


# main program
text = input("Enter text: ")

print("\n=== TEXT ANALYSIS ===")
print("Words:", count_words(text))
print("Vowels:", count_vowels(text))
print("Consonants:", count_consonants(text))
print("Reversed:", reverse_text(text))

if check_palindrome(text):
    print("Palindrome: Yes")
else:
    print("Palindrome: No")

print("Without vowels:", remove_vowels(text))
print("Longest word:", longest_word(text))
print("Word Frequency:", word_frequency(text))
```
