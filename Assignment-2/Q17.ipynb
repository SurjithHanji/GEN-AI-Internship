```python
# Q17 - Palindrome Checker

# taking input from user
s = input("Enter word/number: ")

# converting to lowercase
text = s.lower()

# reversing the text
rev = text[::-1]

# printing original and reversed
print("Original:", s)
print("Reversed:", rev)

# checking palindrome
if text == rev:
    print("Result: PALINDROME")
else:
    print("Result: NOT a palindrome")
```
