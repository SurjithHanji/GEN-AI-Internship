```python
# Q14 - Factorial Calculator

# taking input from user
n = int(input("Enter a number: "))

# checking negative number
if n < 0:
    print("Factorial not defined for negative numbers.")

# checking zero
elif n == 0:
    print("0! = 1")

else:
    fact = 1

    # calculating factorial
    for i in range(1, n + 1):
        fact = fact * i

    # printing result
    print(n, "! =", fact)
```
