```python
# Q6 - Grade Calculator (Any Range Marks)

# taking marks and maximum marks from user
m1 = int(input("Enter marks of subject 1: "))
max1 = int(input("Enter maximum marks of subject 1: "))

m2 = int(input("Enter marks of subject 2: "))
max2 = int(input("Enter maximum marks of subject 2: "))

m3 = int(input("Enter marks of subject 3: "))
max3 = int(input("Enter maximum marks of subject 3: "))

m4 = int(input("Enter marks of subject 4: "))
max4 = int(input("Enter maximum marks of subject 4: "))

m5 = int(input("Enter marks of subject 5: "))
max5 = int(input("Enter maximum marks of subject 5: "))

# finding total obtained marks
total = m1 + m2 + m3 + m4 + m5

# finding total maximum marks
total_max = max1 + max2 + max3 + max4 + max5

# finding percentage
percent = (total / total_max) * 100

# checking pass or fail (40% in each subject)
if (m1/max1)*100 >= 40 and (m2/max2)*100 >= 40 and (m3/max3)*100 >= 40 and (m4/max4)*100 >= 40 and (m5/max5)*100 >= 40:
    result = "Pass"
else:
    result = "Fail"

# deciding grade
if percent >= 90:
    grade = "A+"
elif percent >= 80:
    grade = "A"
elif percent >= 70:
    grade = "B"
elif percent >= 60:
    grade = "C"
elif percent >= 50:
    grade = "D"
else:
    grade = "F"

# printing result
print("\n=== RESULT ===")

print("Total Marks:", total)
print("Total Maximum Marks:", total_max)
print("Percentage:", format(percent, ".2f"), "%")
print("Grade:", grade)
print("Result:", result)
```
