```python
# Q9 - Ticket Pricing System

# taking inputs from user
a = int(input("Enter your age: "))
d = input("Enter day of week: ")
n = int(input("Enter number of tickets: "))

# finding base price based on age
if a < 3:
    p = 0
elif a >= 3 and a <= 12:
    p = 150
elif a >= 13 and a <= 59:
    p = 300
else:
    p = 200

# calculating total before discount
total = p * n

# converting day into lowercase
d = d.lower()

# checking weekend discount (20%)
if d == "friday" or d == "saturday" or d == "sunday":
    dis = 0.20 * total
else:
    dis = 0

# calculating final amount
final = total - dis

# printing result
print("\n=== TICKET DETAILS ===")

print("Base price per ticket: ₹", p)
print("Total without discount: ₹", total)
print("Discount: ₹", dis)
print("Total amount: ₹", final)
```
