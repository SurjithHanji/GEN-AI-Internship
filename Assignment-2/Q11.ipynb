```python
# Q11 - Number Pattern Printer

print("Choose Pattern (1-4)")
print("1. Increasing Numbers")
print("2. Same Number Pattern")
print("3. Decreasing Numbers")
print("4. Pyramid Pattern")

ch = int(input("Enter pattern number: "))
h = int(input("Enter height: "))

if h <= 0:
    print("Height must be positive.")
else:

    print("\n=== PATTERN OUTPUT ===")

    # Pattern 1
    if ch == 1:
        for i in range(1, h + 1):
            for j in range(1, i + 1):
                print(j, end=" ")
            print()

    # Pattern 2
    elif ch == 2:
        for i in range(1, h + 1):
            for j in range(i):
                print(i, end=" ")
            print()

    # Pattern 3
    elif ch == 3:
        for i in range(h, 0, -1):
            for j in range(i, 0, -1):
                print(j, end=" ")
            print()

    # Pattern 4
    elif ch == 4:
        for i in range(1, h + 1):
            for j in range(1, i + 1):
                print(j, end="")
            for j in range(i - 1, 0, -1):
                print(j, end="")
            print()

    else:
        print("Wrong pattern choice!")
```
