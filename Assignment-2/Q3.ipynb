```python
# String Manipulator Program (Beginner Level)

# Taking input from user
s = input("Enter a sentence: ")

# Checking if input is empty
if s == "":
    print("Please enter something!")
else:
    w = s.split()   # splitting sentence into words

    print("\nOutput:")
    print("Original sentence:", s)

    # Length including spaces
    print("Total characters (with spaces):", len(s))

    # Length without spaces
    print("Total characters (without spaces):", len(s.replace(" ", "")))

    # Total words
    print("Total words:", len(w))

    # Changing cases
    print("Uppercase:", s.upper())
    print("Lowercase:", s.lower())
    print("Title case:", s.title())

    # First and last word
    print("First word:", w[0])
    print("Last word:", w[-1])

    # Reverse sentence
    print("Reversed sentence:", s[::-1])
```
