```python
# Q4 - Age Calculator

# importing datetime to get current year
from datetime import datetime

# taking birth year from user
year = int(input("Enter your birth year: "))

# getting current year
present_year = datetime.now().year

# checking if birth year is valid
if year > present_year:
    print("Birth year cannot be greater than current year!")
else:
    # calculating age in years
    age = present_year - year

    # calculating age in different units
    months = age * 12
    days = age * 365        # approximate
    hours = days * 24
    minutes = hours * 60

    # calculating years left to reach 100
    left_100 = 100 - age

    # printing results
    print("\n--- Age Details ---")
    print("Your Age:", age, "years")
    print("Age in Months:", months)
    print("Age in Days (approx):", days)
    print("Age in Hours:", hours)
    print("Age in Minutes:", minutes)

    # checking 100 years condition
    if left_100 > 0:
        print("Years left to reach 100:", left_100)
    else:
        print("You are already 100 years or more!")
```
