```python
# Q16 - Number Guessing Game

import random

while True:

    # generating random number
    num = random.randint(1, 100)

    chances = 7
    win = False

    print("\nGuess the number between 1 and 100")

    while chances > 0:

        guess = int(input("Enter your guess: "))
        chances = chances - 1

        if guess == num:
            print("Correct! You guessed the number.")
            win = True
            break

        elif guess < num:
            print("Too low!")

        else:
            print("Too high!")

        print("Chances left:", chances)

    if win == False:
        print("Game Over! The number was:", num)

    again = input("Play again? (yes/no): ")
    if again.lower() != "yes":
        break
```
