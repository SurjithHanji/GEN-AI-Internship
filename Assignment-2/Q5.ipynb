```python
# Q5 - Bill Splitter

# taking input from user
bill = float(input("Enter total bill amount: "))
people = int(input("Enter number of people: "))
tax = float(input("Enter tax percentage: "))
tip = float(input("Enter tip percentage: "))

# checking for valid input
if bill < 0 or people <= 0:
    print("Invalid bill amount or number of people!")
else:
    # calculating tax amount
    tax_value = (tax / 100) * bill

    # adding tax to bill
    after_tax = bill + tax_value

    # calculating tip on bill after tax
    tip_value = (tip / 100) * after_tax

    # final total amount
    total_amount = after_tax + tip_value

    # amount each person should pay
    per_head = total_amount / people

    # printing bill details
    print("\n--- Bill Details ---")
    print("Subtotal: ₹", round(bill, 2))
    print("Tax Amount: ₹", round(tax_value, 2))
    print("Bill After Tax: ₹", round(after_tax, 2))
    print("Tip Amount: ₹", round(tip_value, 2))
    print("Final Total: ₹", round(total_amount, 2))
    print("Amount Per Person: ₹", round(per_head, 2))
```
