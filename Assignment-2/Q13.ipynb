```python
# Q13 - Sum and Average Calculator

# asking how many numbers
n = int(input("How many numbers? "))

if n <= 0:
    print("Count must be greater than zero.")
else:
    total = 0

    # taking numbers from user
    for i in range(1, n + 1):
        num = float(input("Enter number " + str(i) + ": "))
        total = total + num

        # setting first number as max and min
        if i == 1:
            max_val = num
            min_val = num
        else:
            if num > max_val:
                max_val = num
            if num < min_val:
                min_val = num

    # calculating average
    avg = total / n

    # printing results
    print("\n=== RESULTS ===")
    print("Sum:", total)
    print("Average:", avg)
    print("Maximum:", max_val)
    print("Minimum:", min_val)
```
