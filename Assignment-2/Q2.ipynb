```python
## Simple Calculator Program

# Taking input
x = int(input("Enter first number: "))
y = int(input("Enter second number: "))

# Operations
s = x + y        # Addition
d = x - y        # Subtraction
m = x * y        # Multiplication

# Division and Modulus (check zero)
if y != 0:
    div = x / y
    r = x % y
else:
    div = "Not possible (division by zero)"
    r = "Not possible (division by zero)"

p = x ** y       # Power

# Output
print("\nResults:")
print(x, "+", y, "=", s)
print(x, "-", y, "=", d)
print(x, "*", y, "=", m)
print(x, "/", y, "=", div)
print(x, "%", y, "=", r)
print(x, "^", y, "=", p)
```
