```python
# Q20 - Number System Functions

# factorial
def fact(n):
    if n < 0:
        return "Not defined"
    f = 1
    for i in range(1, n + 1):
        f = f * i
    return f

# prime check
def prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

# fibonacci (nth term)
def fib(n):
    if n <= 0:
        return "Invalid"
    a = 0
    b = 1
    for i in range(n - 1):
        c = a + b
        a = b
        b = c
    return a

# sum of digits
def sum_digits(n):
    total = 0
    n = abs(n)
    for digit in str(n):
        total = total + int(digit)
    return total

# reverse number
def reverse_num(n):
    rev = str(n)[::-1]
    return int(rev)

# armstrong number
def armstrong(n):
    total = 0
    power = len(str(n))
    for digit in str(n):
        total = total + int(digit) ** power
    if total == n:
        return True
    else:
        return False

# gcd
def gcd(a, b):
    while b != 0:
        temp = b
        b = a % b
        a = temp
    return a

# lcm
def lcm(a, b):
    return abs(a * b) // gcd(a, b)

# perfect number
def perfect(n):
    total = 0
    for i in range(1, n):
        if n % i == 0:
            total = total + i
    if total == n:
        return True
    else:
        return False


# menu
while True:

    print("\n=== MATH MENU ===")
    print("1.Factorial 2.Prime 3.Fibonacci 4.SumDigits 5.Reverse")
    print("6.Armstrong 7.GCD 8.LCM 9.Perfect 10.Exit")

    ch = int(input("Enter choice: "))

    if ch == 10:
        break

    if ch == 1:
        n = int(input("Enter number: "))
        print("Factorial:", fact(n))

    elif ch == 2:
        n = int(input("Enter number: "))
        print("Prime:", prime(n))

    elif ch == 3:
        n = int(input("Enter number: "))
        print("Fibonacci:", fib(n))

    elif ch == 4:
        n = int(input("Enter number: "))
        print("Sum of digits:", sum_digits(n))

    elif ch == 5:
        n = int(input("Enter number: "))
        print("Reverse:", reverse_num(n))

    elif ch == 6:
        n = int(input("Enter number: "))
        print("Armstrong:", armstrong(n))

    elif ch == 7:
        a = int(input("Enter first number: "))
        b = int(input("Enter second number: "))
        print("GCD:", gcd(a, b))

    elif ch == 8:
        a = int(input("Enter first number: "))
        b = int(input("Enter second number: "))
        print("LCM:", lcm(a, b))

    elif ch == 9:
        n = int(input("Enter number: "))
        print("Perfect number:", perfect(n))

    else:
        print("Wrong choice!")
```
