```python
# Q10 - Simple ATM Simulator

# starting balance
bal = 10000

while True:

    print("\n=== ATM MENU ===")
    print("1. Check Balance")
    print("2. Deposit")
    print("3. Withdraw")
    print("4. Exit")

    ch = int(input("Enter your choice: "))

    # check balance
    if ch == 1:
        print("Current Balance: ₹", bal)

    # deposit money
    elif ch == 2:
        dep = float(input("Enter amount to deposit: "))

        if dep > 0:
            bal = bal + dep
            print("Deposit successful.")
        else:
            print("Amount must be positive.")

    # withdraw money
    elif ch == 3:
        wd = float(input("Enter amount to withdraw: "))

        if wd <= 0:
            print("Invalid amount.")
        elif bal - wd < 500:
            print("Minimum balance of ₹500 must remain.")
        else:
            bal = bal - wd
            print("Withdrawal successful.")

    # exit option
    elif ch == 4:
        print("Thank you for using ATM.")
        break

    else:
        print("Wrong choice! Please select 1 to 4.")
```
